# <span style="color:Red">Credit card spendings</span>

```python
cc_spendings = pd.read_csv("Downloads/creditcardspendings.csv",header=None,names=["Trx date","Reference","Trx amt"])
```

```python
cc_spendings.head()
```

```python
cc_spendings[cc_spendings["Trx amt"].str.contains("cr")]
```

```python
cc_spendings.drop(79,axis=0,inplace=True)
```

```python
cc_spendings["Reference"].unique()
```

```python
cc_spendings.loc[cc_spendings["Reference"].str.contains("I\.T")]
```

```python
category = {"Groccery":["NTUC","COLD STORAGE"],
            "Shopping":["7-ELEVEN","GUARDIAN","LAZADA","FITLION","CHALLENGER","WEITING","HARVEY NORMAN","UNITY","BEST DENKI","C&P T4"],
            "Food":["VIENNA INT L SEAFOOD","4FINGERS","MALA XIANG GUO","BOON TONG KEE","FOOD","SUBWAY","BBQ","COFFEE","POPEYES","RAMEN","MEIDI-YA","WATAMI","SUSHI","KITCHEN","RESTAU","GRAND COP","MCDONALD","WHOLLY","SMARTRETAIL","GRILLED","SAGE","STARBUCKS","DIN TAI FUNG","PHO STREET","OLD TOWN","AKIMITSU","CANTON"],
            "Transport":["GRAB","EZ-LINK"],
            "Subscription":["SPOTIFY","DATAQUEST","COURSRA"],
            # Need to add a backslash before the '.' sign
            "Clothing":["LEVI","UNIQLO","UNDER ARMOUR","I\.T"],
            "Entertainment":["TIMEZONE","TOG","GOLDEN VILLAGE","GV"],
            "Fitness/Health":["FITNESS","SPA","ANYTIME"],
            "Travel":["TRAVEL"],
            "Other":["PAYMENT PLAN"]}
```

```python
for cat_key in category:
    for cat_value in category[cat_key]:
        cc_spendings.loc[cc_spendings["Reference"].str.contains(cat_value),"Category"] = cat_key
```

<span style="color:orange">**Inspect and verify each category**</span>

```python
cc_spendings[cc_spendings["Category"] == "Clothing"]
```

<span style="color:orange">**Make sure every records are assign with category**</span>

```python
cc_spendings[cc_spendings['Category'].isnull()]
```

<span style="color:orange">**Remove strings and convert to numeric**</span>

```python
cc_spendings["Trx amt"] = cc_spendings["Trx amt"].str.replace("S","")
cc_spendings["Trx amt"] = cc_spendings["Trx amt"].str.replace("$","").astype(float)
```

<span style="color:orange">**Chop out the month from trx date into a new column**</span>

```python
cc_spendings["Month"] = cc_spendings["Trx date"].str.slice(3,6)
```

<span style="color:orange">**Remove months with insufficient data**</span>

```python
cc_spendings.groupby("Month").describe()
```

```python
cc_spendings.drop(cc_spendings[cc_spendings["Month"]=="Jul"].index, axis=0,inplace=True)
```

```python
cc_spendings.drop(cc_spendings[cc_spendings["Month"]=="Oct"].index, axis=0,inplace=True)
```

```python
months = cc_spendings["Month"].unique()
```

```python
cc_spend_by_month = {}
cc_spend_by_cat = {}
for month in months:
    cc_spend_by_month[month] = cc_spendings[cc_spendings["Month"] == month].copy()
    cc_spend_by_cat[month] = cc_spend_by_month[month].groupby(by="Category",as_index=False).agg(sum)
    cc_spend_by_cat[month].sort_values(by="Category",ascending=True,inplace=True)
```

```python
cc_spend_by_cat["Sep"]
```

```python
list(category.keys())
```

```python
fig,ax = plt.subplots()
x = cc_spend_by_cat["Sep"]["Category"]
y = cc_spend_by_cat["Sep"]["Trx amt"]
ax.bar(x,y,0.5)
#ax.set_xticklabels(cc_spend_by_cat["Sep"]["Category"],rotation=90)
ax.set_xticklabels(x,rotation=90)
plt.show()
```

## <span style="color:Purple">Plot bar charts for each month</span>

```python
fig = plt.figure(figsize=(20,40))

months_len = len(months)
for i in range(months_len):
    ax = fig.add_subplot(months_len,2,i+1)
    x = cc_spend_by_cat[months[i]]["Category"]
    y = cc_spend_by_cat[months[i]]["Trx amt"]
    ax.bar(x,y,0.5)
    ax.set_xticklabels(x,rotation=90)
    ax.set_ylim(0,1000)
    #ax.set_yticks(arange(0,1000,100))
    ax.set_title(months[i],fontdict={"fontsize":15,"fontweight":10})
    
plt.show()
```

## <span style="color:Purple">Plot pie charts for each month</span>

```python
fig2 = plt.figure(figsize=(20,45))
months_len = len(months)

for i in range(months_len):
    ax2 = fig2.add_subplot(months_len,2,i+1)
    ax2.pie(cc_spend_by_cat[months[i]]["Trx amt"],labels=cc_spend_by_cat[months[i]]["Category"],autopct='%1.1f%%')
    ax2.set_title(months[i],fontdict={"fontsize":15,"fontweight":10})
    plt.axis('equal')

plt.show()
```
